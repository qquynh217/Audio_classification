```python
import numpy as np
import pandas as pd
import os
```

### PReprocessing for fake-voice recognition task

```python

old_data = 'data/train2.csv'

old_df = pd.read_csv(old_data)
old_df.info()
```

```python
#tạo một dataframe mới với slice_file_name là tên file(bỏ .wav) và class là 0
new_df = pd.DataFrame(columns=['slice_file_name', 'class'])
wav_path = 'data\wav/train'
for i in os.listdir(wav_path):
    print(i)
    if i[0].isdigit():
        #THÊM VÀO DATAFRAME
        new_df = pd.concat([new_df, pd.DataFrame({'slice_file_name': [i], 'class': [0]})], ignore_index=True)
    else: break

new_df.tail(5)
```

```python
#đặt class old_df là 1
old_df['class'] = 1
old_df.head(10)
```

```python

df = pd.concat([old_df, new_df], ignore_index=True)
df.info()
```

```python
df.to_csv('data/train_fake_task.csv', index=False)
```

```python
#lam tương tự với test data
old_data = 'data/test2.csv'
old_df = pd.read_csv(old_data)
new_df = pd.DataFrame(columns=['slice_file_name', 'class'])
wav_path = 'data\wav/test'
for i in os.listdir(wav_path):
    print(i)
    if i[0].isdigit():
        new_df = pd.concat([new_df, pd.DataFrame({'slice_file_name': [i], 'class': [0]})], ignore_index=True)
    else: break
old_df['class'] = 1
df = pd.concat([old_df, new_df], ignore_index=True)
df.to_csv('data/test_fake_task.csv', index=False)
```

### Preprocessing for command detection task

```python
labels = ['True', 'False']
path_wav = 'speech_commands'
#khởi tạo dataframe
df = pd.DataFrame(columns=['slice_file_name', 'class'])
for label in labels:
    path = os.path.join(path_wav, label)
    for wav in os.listdir(path):
        #đưa vào dataframe
        df = pd.concat([df, pd.DataFrame({'slice_file_name': [wav], 'class': [label]})], ignore_index=True)
df.head(10)
```

```python
df.tail(10)
```

```python
df.to_csv('data/train_command.csv', index=False)
```
